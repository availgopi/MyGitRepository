```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
print("No. of the rows in this dataset ", len(dataset))
print("No. of the columns in this dataset ", len(dataset.columns))
```

```python
dataset = pd.get_dummies(dataset, columns=['sex'])
```

```python
dataset = pd.get_dummies(dataset, columns=['smoker'], drop_first=True)
```

```python
dataset.columns
```

```python
#split variable to Independent and dependent variables
Ind_var = dataset[['age', 'sex_female', 'sex_male', 'bmi', 'children', 'smoker_yes']]
dep_var = dataset[['charges']]
```

```python
Ind_var
```

```python
dep_var
```

```python
#split the train data and test data into 1/3 ratio
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(Ind_var, dep_var, test_size = 1/3, random_state=0)
```

```python
Y_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor
```

```python
regressor.fit(X_train, Y_train)
```

```python
weight = regressor.coef_
print("Wieght is : ", weight)
bias = regressor.intercept_
print("Bias is: ", bias)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(Y_test, y_pred)
```

```python
r_score
```
