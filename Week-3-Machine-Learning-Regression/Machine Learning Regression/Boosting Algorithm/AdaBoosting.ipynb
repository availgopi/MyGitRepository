```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,columns=['smoker'], drop_first=True)
```

```python
datasets = pd.get_dummies(dataset)
```

```python
datasets
```

```python
datasets.columns
```

```python
indep=datasets[['age', 'bmi', 'children', 'smoker_yes', 'sex_female', 'sex_male']]
dep=datasets[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#X_train = sc.fit_transform(X_train)
#X_test = sc.fit(X_test) 
```

```python
from sklearn.ensemble import AdaBoostRegressor
regressor = AdaBoostRegressor(n_estimators=1000, loss='exponential')
regressor.fit(X_train, y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
