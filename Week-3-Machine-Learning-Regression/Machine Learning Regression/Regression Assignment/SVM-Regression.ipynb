```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
print("The No. Of rows in dataset", len(dataset))
print("The No. Of rows in dataset", len(dataset.columns))
```

```python
dataset = pd.get_dummies(dataset, columns=['sex'])
```

```python
#datasets=pd.get_dummies(dataset,drop_first=True)
datasets = pd.get_dummies(dataset, columns=['smoker'], drop_first=True)
```

```python
datasets
```

```python
#indep=datasets[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
#dep=datasets[['Profit']]
datasets.columns
```

```python
indep=datasets[['age', 'bmi', 'children', 'sex_female', 'sex_male',
       'smoker_yes']]
dep = datasets[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="linear",C=3000)
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```
